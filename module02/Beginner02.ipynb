```python
# import required libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Generate descriptive statistics for numerical columns
# Provide information like mean, standard deviation, min, max, and quartiles.
CarData.describe()
```

```python
# Extract the car's company's name from the variable 'CarName' into the variable 'CarCompany'

CarData['CarCompany'] = CarData['CarName'].str.split().str[0]

CarData.head()
```

```python
# #checking the unique values under "CarCompany"
CarData['CarCompany'].unique()
```

```python
# replacing the mis-spelling with correct ones

CarData = CarData.replace(to_replace='vokswagen', value='volkswagen', regex=True)
CarData = CarData.replace(to_replace='vw', value='volkswagen', regex=True)
CarData = CarData.replace(to_replace='alfa-romero', value='alfa-romeo', regex=True)
CarData = CarData.replace(to_replace='maxda', value='mazda', regex=True)
CarData = CarData.replace(to_replace='Nissan', value='nissan', regex=True)
CarData = CarData.replace(to_replace='porcshce', value='porsche', regex=True)
CarData = CarData.replace(to_replace='toyouta', value='toyota', regex=True)

# again checking the unique values
CarData.CarCompany.unique()
CarData.head()
```

```python
# In seaborn:

sb.set_theme(rc={'figure.figsize':(20,12)}, style="ticks")
flierprops = dict(marker='D', markerfacecolor='black', markersize=4)
bplot = sb.boxplot(data=CarData, x='CarCompany', y='price', hue='CarCompany', flierprops=flierprops)
bplot.set_xlabel(xlabel='Car Company', fontsize=10)
bplot.tick_params(labelsize=9);
```

```python
# Visualising Categorical variables through boxplots

#sb.set_theme(rc={'figure.figsize':(5,12)}, style="ticks")

figure, axes = plt.subplots(1, 4, figsize=(27, 5));

sb.boxplot(data=CarData, x='fueltype', y='price', hue='fueltype', ax=axes[0], flierprops=flierprops);

sb.boxplot(data=CarData, x='cylindernumber', y='price', hue='cylindernumber', ax=axes[1], flierprops=flierprops);

sb.boxplot(data=CarData, x='enginelocation', y='price', hue='enginelocation', ax=axes[2], flierprops=flierprops);

sb.boxplot(data=CarData, x='doornumber', y='price', hue='doornumber', ax=axes[3], flierprops=flierprops);

```
