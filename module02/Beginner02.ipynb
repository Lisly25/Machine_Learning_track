```python
# Extract the car's company's name from the variable 'CarName' into the variable 'CarCompany'

CarData['CarCompany'] = CarData['CarName'].str.split().str[0]

CarData.head()
```

```python
# #checking the unique values under "CarCompany"
CarData['CarCompany'].unique()
```

```python
# replacing the mis-spelling with correct ones

CarData = CarData.replace(to_replace='vokswagen', value='volkswagen', regex=True)
CarData = CarData.replace(to_replace='vw', value='volkswagen', regex=True)
CarData = CarData.replace(to_replace='alfa-romero', value='alfa-romeo', regex=True)
CarData = CarData.replace(to_replace='maxda', value='mazda', regex=True)
CarData = CarData.replace(to_replace='Nissan', value='nissan', regex=True)
CarData = CarData.replace(to_replace='porcshce', value='porsche', regex=True)
CarData = CarData.replace(to_replace='toyouta', value='toyota', regex=True)

# again checking the unique values
CarData.CarCompany.unique()
CarData.head()
```

```python
#creating a dict 'company price' where key = car's company and value = median of their price

company_list = CarData['CarCompany'].unique()

company_price = {}

for i in company_list:
  company_price[i] = CarData[CarData['CarCompany'] == i]['price'].median()

company_price
```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

bucketed_company_price = {}

for company, price in company_price.items():
    if price < 10000:
        bucketed_company_price[company] = 'low'
    elif price >= 10000 and price < 20000:
        bucketed_company_price[company] = 'medium'
    else:
        bucketed_company_price[company] = 'high'

# mapping the company_price on the 'CarCompany' column in the dataset

CarData['CarCompany'] = CarData['CarCompany'].map(bucketed_company_price)

# checking the column's unique values

CarData.CarCompany.unique()
```

```python
CarData = CarData.drop('car_ID', axis=1)
CarData = CarData.drop('symboling', axis=1)
CarData = CarData.drop('CarName', axis=1)
CarData.head()
```

```python
#creating dummy variable for catwegorical variables
CarData = pd.get_dummies(CarData)

#checking the dataset
CarData.head()
```

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

CarData_train, CarData_test = train_test_split( CarData, train_size = 0.7, test_size = 0.3, random_state = 100)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
```

```python
#importing required function for scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

y_train = CarData_train['price']
X_train = CarData_train.drop(['price'], axis=1)

y_test = CarData_test['price']
X_test = CarData_test.drop(['price'], axis=1)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Making predictions
predicted_price = model.predict(X_test_scaled)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.

sb.set_theme(rc={'figure.figsize':(13,10)}, style="ticks")
sb.scatterplot(x=y_test, y=predicted_price, s=100)
plt.ylim(0, 45000)
plt.title('y_test vs predicted_price', fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xlabel('y_test', fontsize=20)
plt.ylabel('predicted_price', fontsize=20)
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predicted_price)
print(r2)
```
