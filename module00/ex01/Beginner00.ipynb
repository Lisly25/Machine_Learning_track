The goal of the exercise was to improve on the graph from the previous exercise by displaying linear regression

```python
import pandas as pd
import seaborn as sb
```

```python
from google.colab import files
uploaded = pd.read_csv("Salary_dataset.csv")
```

```python
sb.set_theme()
```

```python
dataframe = pd.DataFrame(data=uploaded)
plot = sb.regplot(data=dataframe, x="YearsExperience", y="Salary", color="b")
plot.set_title("Linear Regression: Years of Experience vs Salary")
plot.set_xlabel("Years of Experience");
```
