```python
# Drop the 'index' column
df_cleaned_up = df.drop('index', axis=1)
df_cleaned_up
```

```python
# Training and test data split (70:30)
from sklearn.model_selection import train_test_split
X = df_cleaned_up.drop('type', axis=1)
y = df_cleaned_up['type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2442)
```

```python
# Display class distribution in training and test data
print("Class distribution in training data:")
print(y_train.value_counts())
print("\nClass distribution in test data:")
print(y_test.value_counts())
```

```python
#Create pie chart of distribution in training data
import matplotlib.pyplot as plt
plt.figure(figsize=(9, 9))
plt.pie(y_train.value_counts(), labels=y_train.value_counts().index, autopct='%1.1f%%', startangle=140)
plt.title('Class Distribution in Training Data')
plt.show()
```

```python
# Multinomial logistic regression model
# Training the model
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the model
model = LogisticRegression(max_iter=1000, multi_class='multinomial')
model.fit(X_train_scaled, y_train)
```

```python
# Prediction on the training data
model.fit(X_train_scaled, y_train)
prediction = model.predict(X_train_scaled)
accuracy = model.score(X_train_scaled, y_train)
print("Accuracy on training data:", accuracy)
```

```python
# Prediction on the test data
prediction_on_test = model.predict(X_test_scaled)
accuracy_on_test = model.score(X_test_scaled, y_test)
print("Accuracy on test data:", accuracy_on_test)
```

```python
# Calculate the confusion matrix
from sklearn.metrics import confusion_matrix
cm_train = confusion_matrix(y_train, prediction)
cm_test = confusion_matrix(y_test, prediction_on_test)

print("Confusion matrix on training data:")
print(cm_train)
print("\nConfusion matrix on test data:")
print(cm_test)

# Create a heatmap of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', cbar=False);
```
